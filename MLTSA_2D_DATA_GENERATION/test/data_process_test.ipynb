# Spiral 2-branch

```python
raw_s2 = generate_traj('s2',number=100,visual=True)
```

```python
dp = DataProcess()
dp.eps=2
```

```python
label_s2 = dp.label(raw_s2, visual=True)
plt.title("Labelled Trajectories of 2-Branch Spiral Dataset")
```

```python
cd_s2, cl_s2 = dp.clean(raw_s2, label_s2, cnum=2, visual=True)
plt.title("Clean data after picking the top 2 frequent Clusters")
```

```python
plt.bar(x=0, color='b', height=list(cl_s2).count(0), width=0.5, label="Class0")
plt.bar(x=1, color='r', height=list(cl_s2).count(1), width=0.5, label="Class1")
plt.legend(loc='best')
plt.title("Frequency of different classes before balancing data")
```

```python
bd_s2, bl_s2 = dp.filter(cd_s2, cl_s2, fsize=30, visual=True)
plt.title("Balanced Trajectory Classes")
```

```python
plt.bar(x=0, color='b', height=list(bl_s2).count(0), width=0.5, label="Class0")
plt.bar(x=1, color='r', height=list(bl_s2).count(1), width=0.5, label="Class1")
plt.legend(loc='best')
plt.title("Frequency of different classes after balancing data")
```
